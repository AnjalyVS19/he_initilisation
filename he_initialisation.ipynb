```python
train=data_train.drop('label',axis=1).transpose()
print(train.shape)
train=np.array(train/255.0)
labels=data_train['label'].values.reshape(1,60000)
print(labels.shape)
labels_dev=np.zeros((60000,10))
print(labels_dev[0])
labels_dev[np.arange(60000),labels]=1
print(labels_dev[0])
labels=labels_dev.transpose()
print(labels.shape)
```

```python
# he initialisation, std=2/sqrt(fanin)
```

```python
def create_layer(x,n,activation):
    ndim=int(x.shape[0])
    stddev=2/np.sqrt(ndim)
    init=tf.truncated_normal((n,ndim),stddev=stddev)
    w=tf.Variable(init)
    b=tf.Variable(tf.zeros([n,1]))
    z=tf.matmul(w,x)+b
    return activation(z)
```

```python
n_dim=784
n1=300
n2=300
n_output=10
x=tf.placeholder(tf.float32,[n_dim,None])
y=tf.placeholder(tf.float32,[10,None])
learning_rate=tf.placeholder(tf.float32,shape=())
hidden1=create_layer(x,n1,activation=tf.nn.relu)
hidden2=create_layer(hidden1,n2,activation=tf.nn.relu)
output=create_layer(hidden2,n_output,activation=tf.identity)
y_=tf.nn.softmax(output)
```

```python
cost=tf.reduce_mean(y*tf.log(y_)+(1-y)*tf.log(1-y_))
optimizer=tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
def model(minibatchsize,trainingepochs,features,classes,logging_step=100,learning_r=0.001):
    sess=tf.Session()
    sess.run(tf.global_variables_initializer())
    cost_history=[]
    for epoch in range(trainingepochs+1):
        for i in range(0,features.shape[1],minibatchsize):
            x_trainmini=features[:,i:i+minibatchsize]
            y_trainmini=classes[:,i:i+minibatchsize]
            sess.run(optimizer,feed_dict={x:x_trainmini,y:y_trainmini,learning_rate:learning_r})
        cost_=sess.run(cost,feed_dict={x:features,y:classes})
        cost_history=np.append(cost_history,cost_)
        if(epoch % logging_step)==0:
            print('reached epoch',epoch,'cost',cost_)
    return sess,cost_history
```

```python
sess,cost_history=model(minibatchsize=128,trainingepochs=50,features=train,classes=labels,logging_step=10,learning_r=0.001)
```
